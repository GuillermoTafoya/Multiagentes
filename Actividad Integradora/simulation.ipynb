```python
%%capture
paleta = np.array(
    [[30 , 30 , 30 ],    # 0 -> garage floor
    [ 106, 90 , 84 ],    # 1 -> box
    [ 40 , 44 , 44 ],    # 2 -> stack (0/5)
    [ 81 , 88 , 88 ],    # 3 -> stack (1/5)
    [ 122, 132, 132],    # 4 -> stack (2/5)
    [ 163, 176, 176],    # 5 -> stack (3/5)
    [ 204, 220, 220],    # 6 -> stack (4/5)
    [ 142, 229, 229],    # 7 -> stack (5/5)
    [ 225, 229, 142],    # 8 -> robot with cargo
    [ 97, 102, 15]])   # 9 -> robot without cargo

all_grid = board.datacollector.get_model_vars_dataframe()

pinta_me = lambda matriz: paleta[matriz.astype(int)]

fig, axs = plt.subplots(figsize=(7,7))
axs.set_xticks([])
axs.set_yticks([])
patch = plt.imshow(pinta_me(all_grid.iloc[0][0]))
#plt.colorbar(ticks=np.linspace(0,2,3), fraction=0.046, pad=0.04)

def animate(i):
    patch.set_data(pinta_me(all_grid.iloc[i][0]))
anim = animation.FuncAnimation(fig, animate, frames=len(all_grid))
```
